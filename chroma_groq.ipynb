```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="mixtral-8x7b-32768",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to French. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
ai_msg = llm.invoke(messages)
ai_msg
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate.from_messages([
    ("system", """
                You are a helpful AI bot.
                Your name is {name}.
                Your job is to translates {input_language} to {output_language}.
                Just give one line response. Only translation and nothing more in your response.
                Also mention your name.
                """
    ),
    ("human", "{user_input}"),
])

chain = prompt_template | llm

response = chain.invoke(
                {
                    "name": "Annu",
                    "input_language": "English",
                    "output_language": "Hindi",
                    "user_input": "I love my baby boy!"
                }
            )

response.content
```

```python
import numpy as np
from sentence_transformers import SentenceTransformer

print(np.__version__)

# Define the function for embedding
model = SentenceTransformer("all-MiniLM-L6-v2")

def sentence_transformer_ef(doc):
    return model.encode(doc)

documents = ["This document is about New York", "This document is about Goa"]
ids = ["id1", "id2"]

for doc, id in zip(documents, ids):
    try:
        embedding = sentence_transformer_ef(doc)
        collection.upsert(
            documents=[doc],
            embeddings=[embedding],
            ids=[id],
        )
        print(f"Successfully upserted document with id: {id}")
    except Exception as e:
        print(f"Failed to upsert document with id: {id} due to error: {e}")
```

```python
from langchain_community.document_loaders import WebBaseLoader

url_loader = WebBaseLoader('https://jobs.apple.com/en-us/details/200556125/software-quality-engineer-mobile-apps-retail-engineering?team=CORSV')
page_data = url_loader.load().pop().page_content
page_data
```

```python
import re
data = ' '.join(re.split(r'[\n\t]+', page_data)).strip()
data
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template(
    """
    ### SCRAPPED TEXT FROM WEBSITE: {website_data}
    ### CREATE JSON ABSTRACT with keys: role, experience, job description, skills, pay.
    Only return JSON response and not the preamble.
    Pretty print the JSON response, for example it should not have '\n'
    """
)

chain = prompt_template | llm
result = chain.invoke(
    {
        'website_data':data
    }
)

json = ''.join(result.content.split('\n'))
print(json)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_response = json_parser.parse(json)
json_response
```

```python
import pandas as pd

df = pd.read_csv('my_portfolio.csv')
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('myvectorstore')
my_collection = client.get_or_create_collection(name='portfolio_collection')

# If collections is empty
if not my_collection.count():
    # Iterate over _ which is index and row
    for _, row in df.iterrows():
        embedding = sentence_transformer_ef(row['Techstack'])
        my_collection.add(
            embeddings=embedding,
            ids=[str(uuid.uuid4())],
            documents=row['Techstack'],
            metadatas={'link':row['Links']}
        )
else:
    print('Colletion is not empty')

my_collection.get()
```
